```python
# Carregamento dos dados
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("etapa1_youtube_emojis.csv")
df.head()
```

```python
# Gráfico contendo as distribuições da razão ódio/não ódio no YouTube
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman", "Times", "DejaVu Serif"]

plt.figure(figsize=(8,5), dpi=300)
plt.hist(df["razao_odio_nao_odio"], bins=55, color="#54B420", edgecolor="black", alpha=0.7)
plt.axvline(1, color="red", linestyle="--", label="Razão = 1 (neutra)")

plt.xlim(0.0, 2.0)
plt.xticks([0.0, 0.5, 1.0, 1.5, 2.0])

plt.ylim(0, 50)
plt.yticks([0, 10, 20, 30, 40, 50])

plt.xlabel("Razão Ódio / Não Ódio (suavizada)", fontsize=16)
plt.ylabel("Número de emojis", fontsize=16)
plt.legend()
plt.show()
```
