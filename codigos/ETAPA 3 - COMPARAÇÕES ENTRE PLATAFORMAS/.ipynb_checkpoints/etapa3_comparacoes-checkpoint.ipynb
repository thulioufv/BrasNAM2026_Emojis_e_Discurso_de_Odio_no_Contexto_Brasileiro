```python
# Leitura das bases de dados
import pandas as pd

# Base do Instagram (HateBR)
df_instagram = pd.read_csv(
    "etapa1_instagram_emojis.csv"
)

# Adiciona uma coluna identificando a plataforma
df_instagram["plataforma"] = "Instagram"


# Base do Twitter (ToLD / TuPy)
df_twitter = pd.read_csv(
    "etapa1_twitter_emojis.csv"
)

# Adiciona uma coluna identificando a plataforma
df_twitter["plataforma"] = "Twitter"


# Base do YouTube (OLID-Br)
df_youtube = pd.read_csv(
    "etapa1_youtube_emojis.csv"
)

# Adiciona uma coluna identificando a plataforma
df_youtube["plataforma"] = "YouTube"
```

```python
# Concatenação das bases de dados
# Junta todas as plataformas em um único DataFrame
df_all = pd.concat(
    [df_instagram, df_twitter, df_youtube],
    ignore_index=True
)

df_all.head()
```

```python
# Contagem de emojis por plataforma
# Conta em quantas plataformas cada emoji aparece
contagem_plataformas = (
    df_all.groupby("emoji")["plataforma"]
    .nunique()
    .reset_index(name="n_plataformas")
)

contagem_plataformas.head(20)
```

```python
# Seleciona emojis presentes nas 3 plataformas
emojis_comuns = contagem_plataformas[
    contagem_plataformas["n_plataformas"] == 3
]["emoji"]

# Filtra apenas os emojis comuns
df_comuns = df_all[df_all["emoji"].isin(emojis_comuns)]

# Visualização dos dados filtrados
df_comuns.head(50)
```

```python
# Emojis consistentemente associados ao ódio.
# Para cada emoji, obtém o menor valor da razão entre frequência em ódio e não ódio.
# Isso garante que o emoji tenha comportamento consistente em todas as plataformas.

onsistentes_odio = (
    df_comuns
    .groupby("emoji")["razao_odio_nao_odio"]
    .min()
    .reset_index()
)

# Mantém apenas emojis cuja razão mínima seja maior que 0.20 e ordena do mais associado ao ódio para o menos associado
consistentes_odio = consistentes_odio[
    consistentes_odio["razao_odio_nao_odio"] > 0.20
].sort_values("razao_odio_nao_odio", ascending=False)


# Visualização dos principais emojis
consistentes_odio.head(10)
```

```python
# Emojis divergentes entre plataformas
# São aqueles que possuem comportamento muito diferente entre plataformas,
# podendo estar associados ao ódio em uma rede e ao não ódio em outra.

divergencia = (
    df_comuns
    .groupby("emoji")["razao_odio_nao_odio"]
    .agg(["min", "max"])
    .reset_index()
)

# Calcula a diferença entre o maior e o menor valor da razão em diferentes plataformas
divergencia["amplitude"] = divergencia["max"] - divergencia["min"]

# Ordena pelos emojis mais divergentes
divergencia.sort_values("amplitude", ascending=False).head(10)
```

```python
# Seleciona o emoji com maior divergência

emoji_exemplo = divergencia.sort_values(
    "amplitude", ascending=False
).iloc[0]["emoji"]

df_all[df_all["emoji"] == emoji_exemplo]
```

```python
# Conta quantos emojis diferentes aparecem em cada plataforma

# A menor diversidade de emojis em determinadas plataformas pode sugerir normas discursivas
# implícitas, restringindo o uso de elementos paralinguísticos mesmo quando tecnicamente disponíveis.

df_all.groupby("plataforma")["emoji"].nunique()
```

```python
# Classifica a associação do emoji com base na razão entre frequência em mensagens de ódio e não ódio.

def classificar_razao(r):
    if r > 2:
        return "Fortemente ódio"
    elif r > 1.2:
        return "Levemente ódio"
    elif r < 0.5:
        return "Fortemente não ódio"
    elif r < 0.8:
        return "Levemente não ódio"
    else:
        return "Neutro"

# Aplica a classificação em cada emoji
df_all["classe_associacao"] = df_all["razao_odio_nao_odio"].apply(classificar_razao)

df_all.head()
```

```python
# Gera uma tabela cruzada mostrando a proporção de cada classe de associação em cada plataforma

pd.crosstab(
    df_all["plataforma"],
    df_all["classe_associacao"],
    normalize="index"
)
```
